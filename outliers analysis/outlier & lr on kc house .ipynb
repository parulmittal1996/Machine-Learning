```python
import pandas as pd
import numpy as np
df = pd.read_csv("kc_house_data .csv")
df.head()
```

```python
df.drop(["id","date", "yr_built", "yr_renovated"], axis = 1, inplace = True)

df.head(2)
```

```python
a = df.columns
```

```python
df.shape
```

```python
def outlier(data):
    a = df.columns
    for i in range(len(a)):
        d=data[a[i]]
        outlier1=[]
        data_mean = np.mean(d)
        data_std = np.std(d)
        threshold = 3
        for j in d:
            zscore = (j - data_mean)/data_std
        
            if(np.abs(zscore)> threshold):
                outlier1.append(j)
        for m in outlier1:
            data = data[data[data.columns[i]] !=m]
    return data
```

```python
outlier(df)
```

```python
y = df.iloc[:,0]
x = df.iloc[:,1:18]

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .20, random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
p = LinearRegression()
```

```python
p.fit(x_train,y_train)
```

```python
pre = p.predict(x_test)
```

```python
pre
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test,pre))
```

```python
rmse
```

```python
r2_score(y_test, pre)
```

```python
print('Intercept: \n', p.intercept_)
print('Coefficients: \n', p.coef_)
```
