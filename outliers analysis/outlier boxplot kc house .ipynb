```python
def outlier(data):
    out = []

    q1,q3=np.percentile(data,[25,75])
    l = q1 - (1.5*(q3-q1))
    u = q3 + (1.5*(q3-q1))
    for i in data:
        if(i>=u) or (i<=l):
            out.append(i)
    return(out)
outlier(x)
    
```
