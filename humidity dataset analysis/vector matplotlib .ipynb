```python
import numpy as np
x1 = np.array([0,0,8,15])
x2 = np.array([0,0,5,20])
x3 = x1 + x2
plt.xlim(-15,15)
plt.ylim(-40,40)
plt.quiver(x1[0],x1[1],x1[2],x1[3],scale_units ="xy", angles ="xy", scale = 1, color = "green")
plt.quiver(x2[0],x2[1],x2[2],x2[3],scale_units ="xy", angles ="xy", scale = 1, color = "blue")
plt.quiver(x3[0], x3[1], x3[2], x3[3],scale_units ="xy", angles ="xy", scale = 1, color = "red")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

soa = np.array([[0, 0, 1, 1, -2, 0], [0, 0, 2, 1, 1, 0],
                [0, 0, 3, 2, 1, 0], [0, 0, 4, 0.5, 0.7, 0]])

X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([-1, 0.5])
ax.set_ylim([-1, 1.5])
ax.set_zlim([-1, 8])
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

soa = np.array([[0, 0, 0, 10, 10, 10], [0, 0, 0, 2,6,7]])
x3 = np.array([0,0,0,12,16,17])

X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.quiver(x3[0],x3[1],x3[2],x3[3],x3[4],x3[5],color="red")

ax.set_xlim([0, 12])
ax.set_ylim([0, 16])
ax.set_zlim([0, 17])
plt.show()
```
