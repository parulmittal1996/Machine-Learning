```python
import pandas as pd
data= pd.read_csv("logisticcodedata2.csv")
data.head(2)
```

```python
import matplotlib.pyplot as plt
plt.style.use("default")
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,2]
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],marker='o',c=y)
plt.axis([-2,10,-2,10])
plt.show()
```

```python
from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='linear', C=1E10)
model.fit(x, y)
```

```python
k=model.predict(x)
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],marker='o',c=k)
plt.axis([-2,10,-2,10])
plt.show()
```

```python
import pandas as pd
data= pd.read_csv("svmdata.csv")
data
```
