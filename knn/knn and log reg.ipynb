```python
import pandas as pd
dataset = pd.read_csv("knn_data1.csv")
dataset
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 1/3, random_state = 0 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
k = LogisticRegression()
```

```python
k.fit(x_train,y_train)
```

```python
l = k.predict(x_test)
```

```python
l
```

```python
from sklearn import datasets, linear_model, metrics 
print("Logistic Regression model accuracy(in %):",  
metrics.accuracy_score(y_test, l)*100)
```

```python
from sklearn.metrics import confusion_matrix 
results = confusion_matrix(y_test, l) 
print ('Confusion Matrix :')
print(results)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=15)
neigh.fit(x_train, y_train)
```

```python
n =neigh.predict(x_test)
```

```python
n
```

```python
from sklearn import datasets,neighbors , metrics 
print("  KNeighborsClassifier accuracy(in %):",  
metrics.accuracy_score(y_test,n)*100)
```

```python
from sklearn.metrics import confusion_matrix 
results = confusion_matrix(y_test, n) 
print ('Confusion Matrix :')
print(results)
```
